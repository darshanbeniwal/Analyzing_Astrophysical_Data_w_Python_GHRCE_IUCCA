#Load the reuqired packages

```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

#Estimate value of $\pi$ using Monte Carlo

```python
n = 10000000  # number of random points to generate
count = 0  # count of points that fall within the unit circle

for i in range(n):
    x = random.uniform(-1, 1)  # generate random x-coordinate
    y = random.uniform(-1, 1)  # generate random y-coordinate
    if x**2 + y**2 <= 1:  # check if point falls within the unit circle
        count += 1

pi = 4 * count / n  # calculate pi based on the count of points

print(pi)
```

#*Visualize* the estimation of $\pi$ value using Monte Carlo

```python
n = 1000  # number of random points to generate
count = 0  # count of points that fall within the unit circle

# generate random points and check if they fall within the unit circle
x_inside = []
y_inside = []
x_outside = []
y_outside = []
for i in range(n):
    x = random.uniform(-1, 1)  # generate random x-coordinate
    y = random.uniform(-1, 1)  # generate random y-coordinate
    if x**2 + y**2 <= 1:  # check if point falls within the unit circle
        count += 1
        x_inside.append(x)
        y_inside.append(y)
    else:
        x_outside.append(x)
        y_outside.append(y)

# estimate the value of pi
pi = 4 * count / n

# plot the square and circle
fig, ax = plt.subplots(figsize=(6, 6))
circle = plt.Circle((0, 0), 1, color='r', fill=False)
square = plt.Rectangle((-1, -1), 2, 2, color='b', fill=False)
ax.add_artist(circle)
ax.add_artist(square)

# plot the random points
plt.scatter(x_inside, y_inside, color='g')
plt.scatter(x_outside, y_outside, color='r')

# set axis limits and aspect ratio
ax.set_xlim([-1.1, 1.1])
ax.set_ylim([-1.1, 1.1])
ax.set_aspect('equal', adjustable='box')

# add title and legend
plt.title('Estimating Pi using Monte Carlo')
# plt.legend(['Unit Circle', 'Bounding Square', f'Estimated Pi = {pi:.4f}', 'Points inside Circle', 'Points outside Circle'])

plt.show()
```
